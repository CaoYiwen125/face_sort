# PCA特征提取

导入需要的库，图像及标签训练数据faceR、faceDR、fcaeS、faceDS已经过预处理

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

1、获取数字化后的图像训练数据
数据的第一列为图像编号，为方便降维和使用分类器处理，需要删掉此列

```python
##########获取数字化后的图像训练数据##########
f = open("faceR","r",encoding="utf8")
x = f.read()
f.close()
x = x.split() #默认删除所有空字符，包括空格和换行符等
x = np.array(list(float(char) for char in x)) #字符串列表转数组
x.resize((1996,100)) #转为二维数组
xtrain = x
print('图像训练数据xtrain:\n',xtrain,xtrain.shape)
xtrain = np.delete(xtrain,0,axis=1) #删除第一列：图像编号
print('\nxtrain去编号后：\n',xtrain,xtrain.shape)
```

```python
#标准化图像训练数据
scaler = StandardScaler()
xtrain = scaler.fit_transform(xtrain)
n_features = xtrain.shape[1]
n_samples = xtrain.shape[0]
print('\nxtrain标准化后：\n',xtrain,"\n特征数：",n_features,"\n样本数：",n_samples)
```

2、获取预处理后的标签训练数据

```python
#########获取预处理后的标签训练数据##########
f = open("faceDR","r",encoding="utf8")
y = f.read()
f.close()
y = y.split() #默认删除所有空字符，包括空格和换行符等
#print(y)
del y[0]
#print(y)
y = np.array(y)
y.resize((1996,5))
ytrain = y
print('标签训练数据ytrain:\n',ytrain,ytrain.shape)
```

```python
#性别分类
print('\n性别分类ytrain_sex:\n',ytrain[:,1],ytrain[:,1].shape)
#年龄分类
print('\n年龄分类ytrain_age:\n',ytrain[:,2],ytrain[:,2].shape)
#种族分类
print('\n种族分类ytrain_race:\n',ytrain[:,3],ytrain[:,3].shape)
#表情分类
print('\n表情分类ytrain_face:\n',ytrain[:,4],ytrain[:,4].shape)
```

3、获取预处理后的测试数据

```python
###########获取预处理后的测试数据##########
f = open("faceS","r",encoding="utf8")
x2 = f.read()
f.close()
x2 = x2.split() #默认删除所有空字符，包括空格和换行符等
x2 = np.array(list(float(char) for char in x2)) #字符串列表转数组
x2.resize((1996,100)) #转为二维数组
xtest = x2
print('图像测试数据xtest:\n',xtest,xtest.shape)
xtest = np.delete(xtest,0,axis=1) #删除第一列：图片编号
print('\nxtest去编号后：\n',xtest,xtest.shape)
xtest = scaler.fit_transform(xtest)
print('\nxtest标准化后：\n',xtest)
```

4、获取预处理后的标签测试数据

```python
###########获取预处理后的标签测试数据##########
f = open("faceDS","r",encoding="utf8")
y2 = f.read()
f.close()
y2 = y2.split() #默认删除所有空字符，包括空格和换行符等
#print(y2)
y2 = np.array(y2)
y2.resize((1996,5))
ytest = y2
print('标签测试数据ytest:\n',ytest,ytest.shape)
```

```python
#性别分类
print('\n性别分类ytest_sex:\n',ytest[:,1],ytest[:,1].shape)
#年龄分类
print('\n年龄分类ytest_age:\n',ytest[:,2],ytest[:,2].shape)
#种族分类
print('\n种族分类ytest_race:\n',ytest[:,3],ytest[:,3].shape)
#表情分类
print('\n表情分类ytest_face:\n',ytest[:,4],ytest[:,4].shape)
```

5、创建PCA模型降维并训练

```python
##########创建PCA模型降维并训练###########

#创建性别分类的PCA模型降维并训练
pca_sex = PCA(n_components=1)
pca_sex.fit(xtrain)
#将训练好的PCA模型对训练数据和测试数据进行特征提取
xtrain_sex = pca_sex.transform(xtrain)
xtest_sex = pca_sex.transform(xtest)
print('用训练后的性别分类PCA模型降维后的图像训练数据和测试数据：',xtrain_sex.shape,xtest_sex.shape)
```

```python
#创建年龄分类的PCA模型降维并训练
pca_age = PCA(n_components=2)
pca_age.fit(xtrain)
#将训练好的PCA模型对训练数据和测试数据进行特征提取
xtrain_age = pca_age.transform(xtrain)
xtest_age = pca_age.transform(xtest)
print('用训练后的年龄分类PCA模型降维后的图像训练数据和测试数据：',xtrain_age.shape,xtest_age.shape)
```

```python
#创建种族分类的PCA模型降维并训练
pca_race = PCA(n_components=6)
pca_race.fit(xtrain)
#将训练好的PCA模型对训练数据和测试数据进行特征提取
xtrain_race = pca_race.transform(xtrain)
xtest_race = pca_race.transform(xtest)
print('用训练后的种族分类PCA模型降维后的图像训练数据和测试数据：',xtrain_race.shape,xtest_race.shape)
```

```python
#创建表情分类的PCA模型降维并训练
pca_face = PCA(n_components=2)
pca_face.fit(xtrain)
#将训练好的PCA模型对训练数据和测试数据进行特征提取
xtrain_face = pca_face.transform(xtrain)
xtest_face = pca_face.transform(xtest)
print('用训练后的表情分类PCA模型降维后的图像训练数据和测试数据：',xtrain_face.shape,xtest_face.shape)
```

# Bayes分类器

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder
import time
```

1、性别分类

```python
# 获取性别分类的标签数据
ytrain_sex = ytrain[:, 1]
# 输出编码后的标签数据形状
print('训练数据形状:', ytrain_sex.shape)

# 获取性别分类的测试数据
ytest_sex = ytest[:, 1]
# 输出编码后的测试数据形状
print('测试数据形状:', ytest_sex.shape)
```

```python
# 创建并训练性别分类器
start_time = time.time()

bys_sex = GaussianNB().fit(xtrain_sex, ytrain[:,1]) #默认alpha=1.0，拉普拉斯平滑
# 对性别分类进行预测
bys_sex.predict(xtest_sex)
print('Bayes分类——性别分类分数',bys_sex.score(xtest_sex,ytest[:,1]))

end_time = time.time()
# 打印运行时间
print("性别分类器训练预测的运行时间:", end_time - start_time, "秒")
```

```python
# 对性别分类进行预测
y_pred_sex = bys_sex.predict(xtest_sex)

# 打印分类报告，包括准确率、召回率、F1分数等指标
report = classification_report(ytest[:, 1], y_pred_sex)
print("Bayes分类 - 性别分类报告:\n", report)
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# 使用LabelEncoder将性别分类的标签转换为数值标签
le = LabelEncoder()
ytest_sex_encoded = le.fit_transform(ytest_sex)
sex_predict_encoded = le.transform(y_pred_sex)

# 计算ROC曲线
fpr, tpr, thresholds = roc_curve(ytest_sex_encoded, sex_predict_encoded)

# 计算AUC值
auc_score = auc(fpr, tpr)

# 绘制ROC曲线
plt.figure()
plt.plot(fpr, tpr, label='ROC curve (AUC = %0.2f)' % auc_score)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Sex Prediction(ROC)')
plt.legend(loc='lower right')
plt.show()

print("AUC值：", auc_score)
```

2、年龄分类

```python
# 获取年龄分类的标签数据
ytrain_age = ytrain[:, 2]
# 输出编码后的标签数据形状
print('训练数据形状:', ytrain_age.shape)

# 获取年龄分类的测试数据
ytest_age = ytest[:, 2]
# 输出编码后的测试数据形状
print('测试数据形状:', ytest_age.shape)
```

```python
# 创建并训练年龄分类器
start_time = time.time()

bys_age = GaussianNB().fit(xtrain_age, ytrain[:,2]) #默认alpha=1.0，拉普拉斯平滑
# 对年龄分类进行预测
bys_age.predict(xtest_age)
print('Bayes分类——年龄分类分数',bys_age.score(xtest_age,ytest[:,2]))

end_time = time.time()
# 打印运行时间
print("年龄分类器训练预测的运行时间:", end_time - start_time, "秒")
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize

# 创建LabelEncoder对象
label_encoder = LabelEncoder()

# 对年龄标签进行编码
ytrain_age_encoded = label_encoder.fit_transform(ytrain_age)
ytest_age_encoded = label_encoder.transform(ytest_age)

# 获取类别数量
num_classes = len(label_encoder.classes_)

# 对测试数据进行独热编码
ytest_age_encoded_binary = label_binarize(ytest_age_encoded, classes=range(num_classes))

# 创建并训练年龄分类器
bys_age = GaussianNB().fit(xtrain_age, ytrain[:,2]) #默认alpha=1.0，拉普拉斯平滑
bys_age.predict(xtest_age)

# 获取分类器的决策函数值
bys_age_pred_proba = bys_age.predict_proba(xtest_age)

# 计算每个类别的ROC曲线和AUC值
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(num_classes):
    fpr[i], tpr[i], _ = roc_curve(ytest_age_encoded_binary[:, i], bys_age_pred_proba[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# 绘制多类别的ROC曲线
plt.figure()

# 绘制每个类别的ROC曲线
for i in range(num_classes):
    plt.plot(fpr[i], tpr[i], label='Class {} (AUC = {:.2f})'.format(i, roc_auc[i]))
    print("贝叶斯分类 - 年龄分类AUC值:", roc_auc[i])

# 绘制对角线
plt.plot([0, 1], [0, 1], 'k--')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic - Bayes (Multi-Class)')
plt.legend(loc='lower right')
plt.show()
```

```python
# 建立性别分类的字典，用于存储每个类别的样本
sex_dict = {}

# 遍历训练数据集
for i in range(len(ytrain_sex)):
    sex_label = ytrain_sex[i]
    sample = xtrain_sex[i]
    
    # 如果类别在字典中不存在，则创建对应的键值对
    if sex_label not in sex_dict:
        sex_dict[sex_label] = []
    
    # 将样本添加到对应类别的列表中
    sex_dict[sex_label].append(sample)

# 计算每个类别的均值向量
sex_means = {}
sex_counts = {}
for sex_label, samples in sex_dict.items():
    mean_vector = np.mean(samples, axis=0)
    sex_means[sex_label] = mean_vector
    sex_counts[sex_label] = len(samples)

# 输出类别、样本数量和对应的均值向量
for sex_label, mean_vector in sex_means.items():
    print('类别:', sex_label)
    print('样本数量:', sex_counts[sex_label])
    print('均值向量:', mean_vector)
```

```python
import numpy as np
import math

start_time = time.time()

NUM = 1996  # 选择分类的测试数据个数
num = 1996  # 和 NUM 保持一致
fault = 0  # 判错个数
sex_predict = np.zeros(num, dtype=np.object)  # 存放预测结果

def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2)**2))

for i in range(num):
    min_distance = float('inf')  # 最小距离
    predicted_class = None
    
    for j in range(len(xtrain_age)):
        distance = euclidean_distance(xtrain_sex[j], xtest_sex[i])  # 计算类别代表向量和测试数据的欧几里得距离
        if distance < min_distance:
            min_distance = distance
            predicted_class = ytrain_sex[j]

    sex_predict[i] = predicted_class
    
    if ytest[i, 1] == predicted_class:
        fault = fault
        # print('分类正确')
    else:
        fault += 1
        # print('分类错误')

sex_predict = np.array(sex_predict)
print('性别分类预测结果：\n', sex_predict)
print('总数 %d 的测试数据中出错个数：%d' % (NUM, fault))

accuracy = accuracy_score(ytest_sex, sex_predict)
print("性别分类分数：", accuracy)

end_time = time.time()
# 打印运行时间
print("性别分类器训练预测的运行时间:", end_time - start_time, "秒")
```

## ROC曲线和AUC值：
  ROC曲线（Receiver Operating Characteristic Curve）是一种显示分类器性能的图形。它以假阳率（False Positive Rate）为横轴，以真阳率（True Positive Rate，即召回率）为纵轴，绘制分类器在不同阈值下的表现。
  在ROC曲线上，每个点代表一个分类器在特定阈值下的假阳率和真阳率。通过改变分类阈值，可以获得不同的ROC曲线。ROC曲线越靠近左上角，表示分类器性能越好。
  AUC（Area Under the Curve）是ROC曲线下的面积。AUC值是一个0到1之间的数值，可以衡量分类器的整体性能。AUC值越接近1，表示分类器具有更好的性能。当AUC等于0.5时，表示分类器性能等于随机猜测。
  ROC曲线和AUC值常用于二分类问题，并且在处理不平衡数据集或需要权衡不同分类阈值时特别有用。它们提供了一种全面评估分类器性能的方式，可以帮助选择最佳分类阈值或比较不同分类器的性能。

## F1分数（F1-score）：
  F1分数是精确率（Precision）和召回率（Recall）的调和平均值。精确率表示预测为正例的样本中实际为正例的比例，召回率表示实际为正例的样本中被正确预测为正例的比例。F1分数综合考虑了精确率和召回率，可以评估分类器的整体性能。
  F1分数的取值范围为0到1，其中1表示最佳性能，0表示最差性能。当精确率和召回率同时很高时，F1分数也会很高。F1分数适用于二分类问题，并且在类别不平衡的情况下效果较好。

```python
from sklearn.metrics import f1_score

print('性别分类上的F1评分：%.2f'%f1_score(ytest[:,1],sex_predict,average='weighted'))
print('年龄分类上的F1评分：%.2f'%f1_score(ytest[:,2],age_predict,average='weighted'))
print('种族分类上的F1评分：%.2f'%f1_score(ytest[:,3],race_predict,average='weighted'))
print('表情分类上的F1评分：%.2f'%f1_score(ytest[:,4],face_predict,average='weighted'))
```

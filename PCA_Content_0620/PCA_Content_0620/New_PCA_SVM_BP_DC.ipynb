```python
from sklearn.svm import SVC
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
import time
```

由于训练数据和测试数据属于取值离散的类别型数据（categorical data），因此使用独热编码OneHotEncoder对数据进行预处理

```python
# 创建OneHotEncoder对象
ohe = OneHotEncoder()
```

1、性别分类

```python
# 获取性别分类的标签数据
ytrain_sex = ytrain[:, 1]
# 对训练数据进行独热编码
ytrain_sex_encoded = ohe.fit_transform(ytrain_sex.reshape(-1, 1)).toarray()
# 输出编码后的标签数据形状
print('编码后的训练数据形状:', ytrain_sex_encoded.shape)

# 获取性别分类的测试数据
ytest_sex = ytest[:, 1]
# 对测试数据进行独热编码
ytest_sex_encoded = ohe.transform(ytest_sex.reshape(-1, 1)).toarray()
# 输出编码后的测试数据形状
print('编码后的测试数据形状:', ytest_sex_encoded.shape)
```

```python
# 定义要调优的参数组合
param_grid = {'C': [0.1, 1, 10, 100], 'kernel': ['linear', 'rbf'], 'gamma': [0.1, 1, 10]}
# 实例化SVM模型
svm = SVC()
# 实例化GridSearchCV对象
grid_search = GridSearchCV(estimator=svm, param_grid=param_grid, cv=5)
# 在训练集上进行网格搜索
grid_search.fit(xtrain_sex, ytrain_sex)
# 打印模型最优参数和输出准确率
print("模型最优参数: ", grid_search.best_params_)
print("准确率: ", grid_search.best_score_)
```

```python
# 创建并训练性别分类器
start_time = time.time()

svm_sex = SVC(kernel='rbf', C=0.1)
svm_sex.fit(xtrain_sex, ytrain_sex_encoded[:, 1])
# 对性别分类进行预测
svm_sex.predict(xtest_sex)
print('SVM分类——性别分类分数',svm_sex.score(xtest_sex,ytest_sex_encoded[:, 1]))

end_time = time.time()
# 打印运行时间
print("性别分类器训练预测的运行时间:", end_time - start_time, "秒")
```

```python
# 创建LabelEncoder对象
label_encoder = LabelEncoder()

# 对性别标签进行编码
ytrain_sex_encoded = label_encoder.fit_transform(ytrain_sex)

# 打印编码后的标签数据形状
print('编码后的训练数据形状:', ytrain_sex_encoded.shape)

# 对性别分类进行预测
svm_sex_pred = svm_sex.predict(xtest_sex)

# 将预测结果转换回类别名称
svm_sex_pred_labels = label_encoder.inverse_transform(svm_sex_pred.astype(int))

# 打印分类报告，包括准确率、召回率、F1分数等指标
report = classification_report(ytest_sex, svm_sex_pred_labels)
print("SVM分类 - 性别分类报告:\n", report)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score

# 创建LabelEncoder对象
label_encoder = LabelEncoder()

# 对性别标签进行编码
ytrain_sex_encoded = label_encoder.fit_transform(ytrain_sex)
ytest_sex_encoded = label_encoder.transform(ytest_sex)

# 获取类别名称
class_names = label_encoder.classes_

# 计算分类器的预测概率
svm_sex_pred_proba = svm_sex.predict(xtest_sex)

# 计算ROC曲线的假正例率（FPR）、真正例率（TPR）和阈值
fpr, tpr, thresholds = roc_curve(ytest_sex_encoded, svm_sex_pred_proba)

# 计算AUC值
auc = roc_auc_score(ytest_sex_encoded, svm_sex_pred_proba)

# 绘制ROC曲线
plt.plot(fpr, tpr, label='SVM ROC curve (AUC = {:.2f})'.format(auc))
plt.plot([0, 1], [0, 1], 'k--')  # 对角线
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic - SVM')
plt.legend(loc='lower right')
plt.show()

print("SVM分类 - 性别分类AUC值:", auc)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize

# 创建LabelEncoder对象
label_encoder = LabelEncoder()

# 对表情标签进行编码
ytrain_face_encoded = label_encoder.fit_transform(ytrain_face)
ytest_face_encoded = label_encoder.transform(ytest_face)

# 获取类别数量
num_classes = len(label_encoder.classes_)

# 对测试数据进行独热编码
ytest_face_encoded_binary = label_binarize(ytest_face_encoded, classes=range(num_classes))

# 创建并训练表情分类器
dt_face = DecisionTreeClassifier(criterion='gini', max_depth=5, min_samples_leaf=1)
dt_face.fit(xtrain_face, ytrain_face)

# 获取分类器的决策函数值
dt_face_pred_proba = dt_face.predict_proba(xtest_face)

# 计算每个类别的ROC曲线和AUC值
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(num_classes):
    fpr[i], tpr[i], _ = roc_curve(ytest_face_encoded_binary[:, i], dt_face_pred_proba[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# 绘制多类别的ROC曲线
plt.figure()

# 绘制每个类别的ROC曲线
for i in range(num_classes):
    plt.plot(fpr[i], tpr[i], label='Class {} (AUC = {:.2f})'.format(i, roc_auc[i]))
    print("决策树分类 - 表情分类AUC值:", roc_auc[i])

# 绘制对角线
plt.plot([0, 1], [0, 1], 'k--')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic - DT (Multi-Class)')
plt.legend(loc='lower right')
plt.show()
```
